```python
import requests
import pandas as pd

def get_place_details(api_key, place_name):
    endpoint = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json'
    params = {
        'input': place_name,
        'inputtype': 'textquery',
        'fields': 'rating,user_ratings_total,price_level,name,type,geometry,place_id',
        'key': api_key
    }

    response = requests.get(endpoint, params=params)

    if response.status_code == 200:
        place_data = response.json()
        return place_data
    else:
        return None

api_key = 'XXXXXXX'

# Mendapatkan nama restoran dari file txt
file_path = '/content/hidden gem restaurant.txt'

try:
    with open(file_path, 'r') as file:
        restaurant_names = file.readlines()
        restaurant_names = [name.strip() for name in restaurant_names if name.strip()]  # Membersihkan spasi dan baris kosong
except FileNotFoundError:
    print("File tidak ditemukan.")
    restaurant_names = []

df = pd.DataFrame(columns=['name', 'price_level', 'rating', 'types', 'user_ratings_total', 'place_id','hidden_gem'])
hidden_gem_loc = []

for restaurant_name in restaurant_names:
    place_details = get_place_details(api_key, restaurant_name)

    if place_details and 'candidates' in place_details:
        candidate = place_details['candidates'][0]  # Hanya mengambil satu tempat pertama
        geometry = candidate.get('geometry', {})
        location = geometry.get('location', {})
        hidden_gem_loc.append(location)

        # Menambahkan baris baru ke DataFrame
        new_row = {
            'name': candidate.get('name', ''),
            'price_level': candidate.get('price_level', ''),
            'rating': candidate.get('rating', ''),
            'types': candidate.get('types', []),
            'user_ratings_total': candidate.get('user_ratings_total', ''),
            'place_id': candidate.get('place_id', ''),
            'hidden_gem': "1"
        }

        df = df.append(new_row, ignore_index=True)  # Menambahkan baris baru ke DataFrame

    else:
        print(f"Gagal mendapatkan detail tempat untuk {restaurant_name}.")

df
```

```python
import requests

def cari_restoran(lat, lng):
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    params = {
        'location': f"{lat},{lng}",
        'radius': 1500,  #
        'type': 'restaurant',
        'key': api_key
    }

    response = requests.get(url, params=params)
    data = response.json()
    return data['results']

for koord in hidden_gem_loc:
    restoran = cari_restoran(koord['lat'], koord['lng'])
    for place in restoran:
      if 'rating' in place and place['rating'] is not None:
        if place['rating'] < 4 or place.get('user_ratings_total', 0) > 5000:
              place_details = get_place_details(api_key, place['name'])

              if place_details and 'candidates' in place_details:
                  candidate = place_details['candidates'][0]  # Hanya mengambil satu tempat pertama

                  # Menambahkan baris baru ke DataFrame
                  new_row = {
                      'name': candidate.get('name', ''),
                      'price_level': candidate.get('price_level', ''),
                      'rating': candidate.get('rating', ''),
                      'types': candidate.get('types', []),
                      'user_ratings_total': candidate.get('user_ratings_total', ''),
                      'place_id': candidate.get('place_id', ''),
                      'hidden_gem': "0"
                  }

                  df = df.append(new_row, ignore_index=True)  # Menambahkan baris baru ke DataFrame
                  # print(new_row)

              else:
                  print(f"Gagal mendapatkan detail tempat untuk {restaurant_name}.")
          # print(f"- {place['name']} ({place['rating']} rating, {place.get('user_ratings_total', 0)} total)")
```
